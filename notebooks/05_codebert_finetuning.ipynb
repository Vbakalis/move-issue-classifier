```python
import sys
sys.path.insert(0, 'src')

from codebert_finetune import run_codebert_finetuning
```

```python
# Run CodeBERT fine-tuning
results = run_codebert_finetuning(
    model_name='microsoft/codebert-base',
    train_path='data/processed/train.csv',
    val_path='data/processed/val.csv',
    test_path='data/processed/test.csv',
    epochs=5,
    batch_size=16,
    lr=2e-5,
    max_length=512,
    results_dir='results',
    output_dir='checkpoints',
)
```

```python
# Compare all available results
import json
from pathlib import Path

print(f"{'Model':<35} {'Val Acc':>8} {'Val F1':>8} {'Test Acc':>9} {'Test F1':>8}")
print('-' * 72)

for p in sorted(Path('results').glob('*_metrics.json')):
    with open(p) as f:
        m = json.load(f)
    name = p.stem.replace('_metrics', '')
    if 'val' in m and 'test' in m:
        print(f"{name:<35} {m['val']['accuracy']:>8.4f} {m['val']['macro_f1']:>8.4f} "
              f"{m['test']['accuracy']:>9.4f} {m['test']['macro_f1']:>8.4f}")
    elif 'accuracy' in m:
        print(f"{name:<35} {'—':>8} {'—':>8} "
              f"{m['accuracy']:>9.4f} {m['macro_f1']:>8.4f}")
```

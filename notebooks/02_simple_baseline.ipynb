## Logistic Regression

```python
from simple_baseline import run_simple_baseline

logreg_results = run_simple_baseline(classifier_name='logreg')
logreg_balanced = run_simple_baseline(classifier_name='logreg', class_weight='balanced')
```

## LinearSVC

```python
svm_results = run_simple_baseline(classifier_name='svm')
svm_balanced = run_simple_baseline(classifier_name='svm', class_weight='balanced')
```

## Random Forest

```python
rf_results = run_simple_baseline(classifier_name='rf')
rf_balanced = run_simple_baseline(classifier_name='rf', class_weight='balanced')
```

## Comparison

```python
import pandas as pd

rows = []
for name, res in [
    ('LogReg', logreg_results), ('LogReg (balanced)', logreg_balanced),
    ('SVM', svm_results), ('SVM (balanced)', svm_balanced),
    ('RF', rf_results), ('RF (balanced)', rf_balanced),
]:
    rows.append({
        'Model': f'TF-IDF + {name}',
        'Val Acc': f"{res['val']['accuracy']:.4f}",
        'Val F1': f"{res['val']['macro_f1']:.4f}",
        'Test Acc': f"{res['test']['accuracy']:.4f}",
        'Test F1': f"{res['test']['macro_f1']:.4f}",
    })
comparison = pd.DataFrame(rows)
print(comparison.to_string(index=False))


import json
from pathlib import Path

print('\n\n=== All Saved Results ===')
print(f"{'Model':<35} {'Val Acc':>8} {'Val F1':>8} {'Test Acc':>9} {'Test F1':>8}")
print('-' * 72)
for p in sorted(Path('results').glob('*_metrics.json')):
    with open(p) as f:
        m = json.load(f)
    name = p.stem.replace('_metrics', '')
    if 'val' in m and 'test' in m:
        print(f"{name:<35} {m['val']['accuracy']:>8.4f} {m['val']['macro_f1']:>8.4f} "
              f"{m['test']['accuracy']:>9.4f} {m['test']['macro_f1']:>8.4f}")
```
